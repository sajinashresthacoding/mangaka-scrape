```python
def mk_request(url):
    '''
    Takes the url and gives you status and returns responses
    '''
    response = requests.get(url)
    if 200 <= response.status_code < 400:
        return response
    else:
        print(f"request returned {response.status_code}error")
```

```python
## let's try scraping all the links for winners
broken_url = []
counter = 1
total_links = len(urls)
manga = []
mangaka = []
year = []

for url in urls:
    try:
        print (f"Scraping year {counter} of {total_links}")
        counter+= 1
        code = scrapes(url)      
        books = code.find_all("h3", class_="p-award-single__title")
        for book in books:
            print(book.get_text())
            manga.append(book.get_text())
            year.append(url)

        authors = code.find_all("div", class_="p-award-single__sub-text")
        for author in authors:
            print(author.get_text())
            mangaka.append(author.get_text())    
    except:
        print(f"{url} was busted: ")
        broken_url.append(f"{url}")
    finally:
        print("*************************************************")
print("Done scraping all units, please check your work")
```
