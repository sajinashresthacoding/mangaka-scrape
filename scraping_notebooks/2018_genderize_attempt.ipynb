```python
def gender_data(name, api_key):
    '''
    Takes a string name and api key and returns gender data using the Genderize API.
    '''
    genderize = Genderize(api_key=api_key)
    gender_data = genderize.get([name])
    return gender_data
```

```python
def gender(name, api_key):
    '''
    takes a string name and determine the gender of the name
    '''
    count = gender_data(name, api_key)[0].get("gender")
    return count
```

```python
def gender_count(name, api_key):
    '''
    takes a string name and determine the count of the name
    '''
    count = gender_data(name, api_key)[0].get("count")
    return count
```

```python
def gender_prob(name, api_key):
    '''
    takes a string name and determine the probability of the gender of the name
    '''
    count = gender_data(name, api_key)[0].get("probability")
    return count
```
