```python
def mk_request(url):
    '''
    Takes the url and gives you status and returns responses
    '''
    response = requests.get(url)
    if 200 <= response.status_code < 400:
        return response
    else:
        print(f"request returned {response.status_code}error")
```

```python
##making soup
def mk_soup(response):
    '''
    makes soup out of the response you give
    '''
    return BeautifulSoup(response.text, "html.parser")
```

```python
## mother function that makes request and soup
def scrapes(url):
    '''
    give it a url to return docsoup of a page
    '''
    return mk_soup(mk_request(url))
```

```python
## gathering the first half of 2018 urls
base18_start = "https://yattatachi.com/"
base18_end = "-2018-manga-release"
```

```python
##gathering the second half of 2018 urls
last18_start = "https://yattatachi.com/"
last18_end = "-2018-manga-light-novel-book-releases"
```

```python
months = ['May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'];
last18_urls = []

for month in months:
    print(f"{last18_start}{month}{last18_end}")
    last18_urls.append(f"{last18_start}{month}{last18_end}")
```

```python
## let's try scraping just one element in 2018 link and see what happens

broken_urls = []
counter = 1
total_links = len(all18_urls)
manga = []
company = []
mangaka = []
medium = []
date = []
urls_month = []

for url in all18_urls:
    try: 
        print (f"Scraping month {counter} of {total_links}")
        counter+= 1
        code = scrapes(url)      
        books = code.find_all("a", class_="release-link")
        for book in books:
            print(book.get_text())
            manga.append(book.get_text())
            urls_month.append(url)
        
        authors = code.find_all("span", class_="release-author")
        for author in authors:
            print(author.get_text())
            mangaka.append(author.get_text()) 
            
        pubs = code.find_all("span", class_="release-company")
        for pub in pubs:
            print(pub.get_text())
            company.append(pub.get_text())
            
        modes = code.find_all("span", class_="release-type")
        for mode in modes:
            print(mode.get_text())
            medium.append(mode.get_text())
            
    except:
        print(f"{url} was busted: ")
        broken_urls.append(f"{url}")
    finally:
        print("*************************************************")
print("Done scraping all units, please check your work")
```
