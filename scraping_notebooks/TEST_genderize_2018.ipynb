```python
##cleaning up dataframe
##1. deleteing "by" from every row of mangaka
df["mangaka"] = df["mangaka"].str.replace("By " , "")
df
```

```python
##2. seperating author from mangaka
##couldnt figure that out so i just seperated in by author first name, i only need that anyways
df['author_firstname'] = df['mangaka'].str.split(' ').str[0]
df
```

```python
## renaming a column cuase it would get confusing 
df.rename(columns = {'name':'book'}, inplace = True)
df
```

```python
##re-ordering columns
df = df[['book', 'mangaka', 'author_firstname', 'company', 'medium', 'month']].copy()
df
```

```python
##using genderize now
## FUNCTION to get gender data from genderize
def gender_data(name):
    '''
    takes a string name and returns gender data
    '''
    converted_to_list = [name]
    gender_data = gd().get(converted_to_list)
    return gender_data
```

```python
def gender_estimate(name):
    '''
    takes a string name, returns gender based on using gender_data function
    '''
    return gender_data(name)[0].get("gender")
```

```python
def gender_prob(name):
    '''
    takes a string name, returns gender probability based on using gender_data function
    '''
    return gender_data(name)[0].get("probability")
```

```python
def gender_certainty(name, min_prob):
    '''
    takes a string name and the min_prob, returns gender probability based on using gender_data function
    '''
    prob = gender_data(name)[0].get("probability")
    if prob >= min_prob:
        return "pass"
    else:
        return "flag"
```

```python
##let's make the occurence

def gender_count(name, appearance):
    '''
    takes a string name and the min_prob, returns gender probability based on using gender_data function
    '''
    count = gender_data(name)[0].get("count")
    if count >= appearance:
        return "sufficient"
    else:
        return "recheck"
```

```python
df["gender"] = df["author_firstname"].apply(lambda x: gender_estimate(x))
df
```
